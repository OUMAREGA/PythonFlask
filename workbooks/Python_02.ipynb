Tranformer ces deux listes en series et nommez-les series1 et series2 :
 - Multipliez-les entre elle pour faire une troisième series
 - Combinez-les pour en faire un dataframe

```python
liste1 = list(range(0, 14))
liste2 = list(range(10, 24))
#series1 = pd.Series(liste1, name="S1")
series1 = pd.Series(liste1)
series2 = pd.Series(liste2)
s3 = series1 * series2
df_fusion = pd.DataFrame({'liste1': series1.values , 'liste2': series2.values, 'liste3': s3.values})
df_fusion
```

En partant des series1 et series2, faites 2 series qui comportent uniquement les valeurs:
   - presentes dans la series1 et non dans la series2
   - presentes dans la series2 et non dans la series1

```python
series3 = [ item for item in series1 if series1.isin(series2)[item] == False ] 
series3

series4 = [ item for item in series2 if series2.isin(series1)[item-10] == False ] 
series4

series1[-series1.isin(series2)]
series2[-series2.isin(series1)]
```
